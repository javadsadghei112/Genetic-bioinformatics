```python
from Bio import Entrez, SeqIO
import pandas as pd
from Bio.Seq import Seq
import matplotlib.pyplot as plt
from Bio import motifs
```

```python
Entrez.email = "enter your email"
handle = Entrez.esearch(db="nucleotide", term='CRT[Gene Name] AND "Plasmodium falciparum"[Organism]', retmax="60")
rec_list = Entrez.read(handle)
handle.close()
```

```python
id_list = rec_list['IdList']
id_list
```

```python
records = []
for id in id_list:
    handle = Entrez.efetch(db="nucleotide", id=id, rettype="gb", retmode="xml")
    record = Entrez.read(handle)
    records.append(record)
    handle.close()

records[0]
```

preprosesing

```python
data = []
for record in records:
    for item in record:
        accession = item['GBSeq_primary-accession']
        description = item['GBSeq_definition']
        dna=item['GBSeq_sequence']

         #dbxrefs = item.get('GBSeq_dbxrefs', None)
        data.append({
            'Accession': accession,
            'Description': description,
            'dna':dna,

            #'DBXrefs': dbxrefs
        })

```

```python
df = pd.DataFrame(data)
df
```

```python
df['Protein'] = df['dna'].apply(lambda x: str(Seq(x).translate()))
df
```

```python
def filter_proteins(df):
    return df[df['Protein'].apply(lambda x: len(x) >= 20)]
filtered_df = filter_proteins(df)
df.shape
```

```python
df['rna'] = df['dna'].apply(lambda x: str(Seq(x).transcribe()))

df
```

```python




df["Length"] = df["dna"].apply(len)

df_sorted = df.sort_values(by="Length", ascending=True).reset_index(drop=True)

df_sorted
```

```python
df.to_csv('nucleotide,Plasmodium falciparum"[Organism].csv')
```
